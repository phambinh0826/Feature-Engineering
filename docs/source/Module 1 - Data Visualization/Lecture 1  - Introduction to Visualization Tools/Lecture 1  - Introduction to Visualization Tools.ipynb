**Lưu ý**: Các kiểu dữ liệu chính được lưu trữ trong các đối tượng *pandas* là `float`, `int`, `bool`, `datetime64[ns]`, `datetime64[ns, tz]`, `timedelta[ns]`, `category`, và `object` (chuỗi). Ngoài ra, các kiểu dữ liệu này còn có kích thước phần tử, ví dụ `int64` và `int32`.


Hãy làm sạch bộ dữ liệu để loại bỏ một số cột không cần thiết. Chúng ta có thể sử dụng phương thức `drop()` của *pandas* như sau:

````python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
````

Hãy đổi tên các cột sao cho có ý nghĩa. Chúng ta có thể sử dụng phương thức `rename()` bằng cách truyền vào một từ điển chứa tên cũ và tên mới như sau:


````python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
````

Chúng ta cũng sẽ thêm một cột `Total` để tính tổng số người nhập cư theo quốc gia trong suốt khoảng thời gian từ 1980 - 2013, như sau:

````python
df_can.columns = df_can.columns.astype(str)  # Chuyển đổi tên cột thành chuỗi
df_can['Total'] = df_can.loc[:, '1980':'2013'].sum(axis=1)
````

Chúng ta có thể kiểm tra xem bộ dữ liệu có bao nhiêu đối tượng `null` như sau:

````python
df_can.isnull().sum()
````

### Lọc Dữ Liệu Dựa trên Điều Kiện

**Biểu đồ đường là gì và tại sao lại sử dụng nó?**

Biểu đồ đường (line chart) hay biểu đồ đường thẳng là một loại biểu đồ hiển thị thông tin dưới dạng một chuỗi các điểm dữ liệu gọi là 'markers' được kết nối với nhau bằng các đoạn thẳng. Đây là một loại biểu đồ cơ bản phổ biến trong nhiều lĩnh vực.
Sử dụng biểu đồ đường khi bạn có một bộ dữ liệu liên tục. Đây là loại biểu đồ phù hợp nhất cho các trực quan hóa dữ liệu dựa trên xu hướng trong một khoảng thời gian.

**Hãy bắt đầu với một nghiên cứu tình huống:**

Vào năm 2010, Haiti đã trải qua trận động đất mạnh 7.0 gây ra thảm họa lớn. Trận động đất đã gây ra sự tàn phá rộng lớn và mất mát về sinh mạng, khoảng ba triệu người đã bị ảnh hưởng bởi thảm họa tự nhiên này. Như một phần trong nỗ lực nhân đạo của Canada, Chính phủ Canada đã tăng cường nỗ lực tiếp nhận người tị nạn từ Haiti. Chúng ta có thể nhanh chóng trực quan hóa nỗ lực này bằng cách sử dụng biểu đồ `Line`:

**Câu hỏi:** Vẽ biểu đồ đường của dữ liệu nhập cư từ Haiti bằng `df.plot()`.

Đầu tiên, chúng ta sẽ trích xuất dữ liệu series cho Haiti.

````python
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
````

Tiếp theo, chúng ta sẽ vẽ biểu đồ đường bằng cách thêm `.plot()` vào dataframe `haiti`.


````python
haiti.plot()
````

*pandas* tự động điền trục x với các giá trị chỉ mục (các năm), và trục y với các giá trị cột (dân số). Tuy nhiên, lưu ý rằng các năm không được hiển thị vì chúng có kiểu dữ liệu là *string*. Do đó, hãy thay đổi kiểu của các giá trị chỉ mục thành *integer* để vẽ biểu đồ.

Ngoài ra, chúng ta hãy gán nhãn cho trục x và trục y bằng `plt.title()`, `plt.ylabel()`, và `plt.xlabel()` như sau:

````python
haiti.index = haiti.index.map(int) # let's change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure
````

Chúng ta có thể nhận thấy rõ ràng sự gia tăng số lượng người nhập cư từ Haiti từ năm 2010 khi Canada tăng cường nỗ lực tiếp nhận người tị nạn từ Haiti. Hãy chú thích sự gia tăng này trên biểu đồ bằng cách sử dụng phương thức `plt.text()`.


````python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show() 
````

Với chỉ vài dòng mã, bạn đã có thể nhanh chóng xác định và trực quan hóa sự gia tăng trong nhập cư!

Lưu ý nhanh về các giá trị x và y trong `plt.text(x, y, label)`:

- Vì trục x (năm) có kiểu dữ liệu là 'integer', chúng ta chỉ định x là một năm. Trục y (số lượng người nhập cư) có kiểu dữ liệu là 'integer', vì vậy chúng ta chỉ cần chỉ định giá trị y = 6000.

```python
    plt.text(2000, 6000, '2010 Earthquake') # năm được lưu dưới dạng kiểu int
```

- Nếu các năm được lưu dưới dạng kiểu 'string', chúng ta sẽ cần chỉ định x là vị trí chỉ mục của năm. Ví dụ, chỉ mục 20 là năm 2000 vì đây là năm thứ 20 với năm gốc là 1980.

```python
    plt.text(20, 6000, '2010 Earthquake') # năm được lưu dưới dạng kiểu int
```

- Chúng ta sẽ học các phương pháp chú thích nâng cao trong các module sau.


Chúng ta có thể dễ dàng thêm nhiều quốc gia vào biểu đồ đường để thực hiện các so sánh có ý nghĩa về số lượng người nhập cư từ các quốc gia khác nhau.

**Câu hỏi:** Hãy so sánh số lượng người nhập cư từ Ấn Độ và Trung Quốc từ năm 1980 đến 2013.

````python
### type your answer here
df_CI = df_can.loc[['India', 'China'], years]
df_CI
````

Bước 2: Vẽ biểu đồ. Chúng ta sẽ chỉ định rõ ràng biểu đồ đường bằng cách truyền tham số `kind` vào phương thức `plot()`.


````python
### type your answer here
df_CI.plot(kind='line')
````

Điều đó có vẻ không đúng...

Nhớ lại rằng *pandas* vẽ các chỉ mục trên trục x và các cột như các đường riêng biệt trên trục y.

Vì `df_CI` là một dataframe với `country` là chỉ mục và `years` là các cột, chúng ta phải chuyển vị dataframe trước bằng cách sử dụng phương thức `transpose()` để hoán đổi hàng và cột.


````python
df_CI = df_CI.transpose()
df_CI.head()
````

*pandas* sẽ tự động vẽ hai quốc gia trên cùng một biểu đồ. Hãy vẽ biểu đồ mới của dataframe đã được hoán đổi. Đảm bảo thêm tiêu đề cho biểu đồ và nhãn cho các trục.

````python
### type your answer here
df_CI.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()


````

Từ biểu đồ trên, chúng ta có thể nhận thấy rằng Trung Quốc và Ấn Độ có xu hướng nhập cư rất giống nhau qua các năm.

**Lưu ý**: Tại sao chúng ta không cần phải hoán đổi dataframe của Haiti trước khi vẽ (như đã làm với df_CI)?

Điều này là vì `haiti` là một series thay vì một dataframe, và có các năm làm chỉ mục như đã được hiển thị dưới đây.

Biểu đồ đường là một công cụ hữu ích để hiển thị nhiều biến phụ thuộc so với một biến độc lập. Tuy nhiên, khuyến nghị rằng không nên có quá 5-10 đường trên một biểu đồ; nếu có nhiều hơn, sẽ trở nên khó khăn để giải thích.

**Câu hỏi:** So sánh xu hướng của 5 quốc gia hàng đầu đóng góp nhiều nhất vào nhập cư vào Canada.


````python
#Step 1: Get the dataset. Recall that we created a Total column that calculates cumulative immigration by country. 
#We will sort on this column to get our top 5 countries using pandas sort_values() method.
    
inplace = True # paramemter saves the changes to the original df_can dataframe
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head(5)

# transpose the dataframe
df_top5 = df_top5[years].transpose() 

print(df_top5)


#Step 2: Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size



plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
````

Chúc mừng! Bạn đã học cách xử lý dữ liệu với Python và tạo một biểu đồ đường với Matplotlib. Có rất nhiều kiểu biểu đồ khác ngoài biểu đồ đường mặc định, tất cả đều có thể được truy cập bằng cách truyền từ khóa `kind` vào phương thức `plot()`. Danh sách đầy đủ các loại biểu đồ có sẵn như sau:
